```python
import numpy as np
import pandas as pd
```

```python
titanic = pd.read_csv('train.csv', sep = ',')
```

# Десятеро найстаріших людей на Титаніку

```python
titanic.sort_values(by= 'Age', ascending = False).head(10)
```

# Кількість людей, які вижили та ні

```python
titanic.Survived.value_counts()
```

# Список чоловіків, котрі були з сім'ями та здійснили посадку у Квінсі

```python
titanic[(titanic.Sex == 'male') & (titanic.Embarked == 'Q') & (titanic.SibSp >= 1)]
```

# Список жінок, котрі вижили

```python
titanic[(titanic.Sex == 'female') & (titanic.Survived == 1)]
```

# Список жінок з 1 класу, котрі вижили

```python
titanic[(titanic.Sex == 'female') & (titanic.Survived == 1) & (titanic.Pclass == 1)].sort_values(by= 'Age')
```

# Відношення врятованих дітей, відносно класу квитка

```python
kids_survived = titanic[(titanic.Age < 19) & (titanic.Survived == 1) & (titanic.Pclass == 1)].sort_values(by= 'Age').shape[0]
```

```python
kids_nsurvived = titanic[(titanic.Age < 19) & (titanic.Survived == 0) & (titanic.Pclass == 1)].sort_values(by= 'Age').shape[0]
```

```python
print('Відношення врятованих дітей 1 класу, до неврятованих:', kids_survived/kids_nsurvived)
```

```python
kids_survived3 = titanic[(titanic.Age < 19) & (titanic.Survived == 1) & (titanic.Pclass == 3)].sort_values(by= 'Age').shape[0]
```

```python
kids_nsurvived3 = titanic[(titanic.Age < 19) & (titanic.Survived == 0) & (titanic.Pclass == 1)].sort_values(by= 'Age').shape[0]
```

```python
print('Відношення врятованих дітей 3 класу, до неврятованих:', kids_survived3/kids_nsurvived3)
```
