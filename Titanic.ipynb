```python
import pandas as pd
train = pd.read_csv('train.csv', sep = ',')
train.head(25)
```

# Жінки, що подорожували другим класом.

```python
train.dropna(inplace = True)
df = pd.DataFrame(train)

input_col_1 = str(input("Enter column name 1: "))
value1 = int(input("Enter a value for searching in col_1: "))
input_col_2 = str(input("Enter column name 2: "))
value2 = str(input("Enter a value for searching in col_2: "))

df.loc[(df[f'{input_col_1}'] == value1) & (df[f'{input_col_2}'] == value2)]
```

# Чоловіки, які старші 30 років та вижили.

```python
train.dropna(inplace = True)
df = pd.DataFrame(train)

input_col_1 = str(input("Enter column name 1: "))
value1 = str(input("Enter a value for searching in col_1: "))
input_col_2 = str(input("Enter column name 2: "))
value2 = int(input("Enter a value for searching in col_2: "))
input_col_3 = str(input("Enter column name 3: "))
value3 = int(input("Enter a value for searching in col_3: "))

df.loc[(df[f'{input_col_1}'] >= value1) & (df[f'{input_col_2}'] >= value2) & (df[f'{input_col_3}'] == value3)]
```

# Люди до 20 років, що заплатили більше 100.

```python
train.dropna(inplace = True)
df = pd.DataFrame(train)

input_col_1 = str(input("Enter column name 1: "))
value1 = int(input("Enter a value for searching in col_1: "))
input_col_2 = str(input("Enter column name 2: "))
value2 = int(input("Enter a value for searching in col_2: "))

df.loc[(df[f'{input_col_1}'] <= value1) & (df[f'{input_col_2}'] >= value2)]
```
